```python
import pandas as pd

import unzip
```

```python
df = pd.read_csv('table.csv')
df
```

```python
unique_ids = df['id'].value_counts()
unique_ids
```

```python
three_times = unique_ids[unique_ids == 3]
print(f"The following values were encountered three times: {three_times.index.tolist()}")
```

```python
repeats = unique_ids.value_counts()

results = zip(repeats.index, repeats.values)
for result in results:
    times, unique_ids = result
    print(f"Here's how many ids repeat {times} times: {unique_ids}")
    
```
